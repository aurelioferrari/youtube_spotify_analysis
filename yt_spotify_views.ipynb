```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Spotify_Youtube.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Likes'] = df['Likes'].fillna(0)
```

```python
df['Comments'] = df['Comments'].fillna(0)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
album_type_count = df['Album_type'].value_counts()
album_type_count
```

```python
labels = album_type_count.index.tolist()
sizes = album_type_count.values.tolist()
plt.rcParams["figure.figsize"] = [8, 8]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Album Types')
plt.legend(labels, loc='upper left')

plt.show()
```

```python
pd.options.display.float_format = '{:.2f}'.format
pd.DataFrame(df.groupby(['Album_type'])['Views'].sum().sort_values(ascending=False))
```

```python
df.groupby('Artist')[['Views', 'Stream']].sum().sort_values(by=['Views', 'Stream'], ascending=False).head(10)
```

```python
df_views = pd.DataFrame(df.groupby(['Artist'])['Views'].sum().sort_values(ascending=False)[:10])
df_views
```

```python
df_stream = pd.DataFrame(df.groupby(['Artist'])['Stream'].sum().sort_values(ascending=False)[:10])
df_stream
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17,6))
ax1.set_title('Top 10 Artists on YouTube')
ax1.ticklabel_format(style='plain')
df_views.plot(kind='bar', ax=ax1, color='darkred')

ax2.set_title('Top 10 Artists on Spotify')
ax2.ticklabel_format(style='plain')
df_stream.plot(kind='bar', ax=ax2, color='darkgreen')


plt.show()
```

```python
## top 10 songs in views
top10_views = pd.DataFrame(df.groupby(['Track'])['Views'].sum().sort_values(ascending=False)[:10])
top10_views
```

```python
## top 10 songs in likes
top10_likes = pd.DataFrame(df.groupby(['Track'])['Likes'].sum().sort_values(ascending=False)[:10])
top10_likes
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17,6))
ax1.set_title('Top 10 Songs by views')
ax1.ticklabel_format(style='plain')
top10_views.plot(kind='bar', ax=ax1, color='lightgreen')

ax2.set_title('Top 10 Songs in Likes')
ax2.ticklabel_format(style='plain')
top10_likes.plot(kind='bar', ax=ax2, color='lightblue')

plt.show()
```

```python
artist_likes = pd.DataFrame(df.groupby(['Artist'])['Likes'].sum().reset_index().sort_values(by='Likes', ascending=False)[:10])
```
